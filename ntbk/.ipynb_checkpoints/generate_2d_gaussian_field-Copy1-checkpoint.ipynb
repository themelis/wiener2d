# This notebook will: 
## 1. Generate a real, Gaussian random field with a given power spectrum
## 2. Demonstrate that the signal covariance is diagonal in Fourier space
## 3. Add noise with variance (heteroscedastic) and mask some pixels (infinite noise)
## 4. Demonstrate that the noise covariance is diagonal in pixel space
## 5. Show that that the "messenger field" covariance is diagonal in both pixel and Fourier space
## 6. Implement messenger field Wiener filter algorithm (https://arxiv.org/abs/1210.4931)_
## 7. Draw samples from the Wiener posterior distribution (https://arxiv.org/abs/1402.1763)

## (These methods avoid needing to invert dense matrices)

```python
%matplotlib inline 

from __future__ import division, print_function
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import LogNorm
cmap = plt.get_cmap('inferno')
plt.rcParams['figure.figsize'] = (12, 4)
```

```python
def radial_profile(data):
    center = data.shape[0]/2
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center)**2 + (y - center)**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile 
```

# Generate a map from a power spectrum

```python
def power_function(k, sigmasignal = 7.5, amplitude = 1e6 ):
    function =  np.exp(-float(k)*float(k)/(2.0*sigmasignal*sigmasignal))/(np.sqrt(2.0*np.pi)*(sigmasignal))
    if function <= 0.0:
        return 1e-19
    else:
        return amplitude*function

vector_power_function = np.vectorize(power_function)
    
_ = plt.subplot(1,2,1), plt.plot(np.linspace(0.0,25.0,40), vector_power_function(np.linspace(0.0,25.0,40)))
_ = plt.ylabel(r'$P(k)$'), plt.xlabel(r'$k$')
# _ = plt.ylim([0.0,0.06])
_ = plt.subplot(1,2,2), plt.plot(np.linspace(0.0,25.0,40), vector_power_function(np.linspace(0.0,25.0,40)))
_ = plt.ylabel(r'$P(k)$'), plt.xlabel(r'$k$')
_ = plt.xscale('log'), plt.yscale('log')
```

```python
size = 128
```

```python
power_map = np.zeros((size, size), dtype = float)
k_map =  np.zeros((size, size), dtype = float)

for (i,j), val in np.ndenumerate(power_map):
    
    k1 = i - size/2.0
    k2 = j - size/2.0
    k_map[i, j] = (np.sqrt(k1*k1 + k2*k2))
    
    if k_map[i,j]==0:
        print(i,j)
        power_map[i, j] = 1e-15
    else:
        power_map[i, j] = power_function(k_map[i, j], 10.0)
```

# Check Fourier signal covariance is diagonal

```python
def make_map(size):
    power_map = np.zeros((size, size), dtype = float)
    k_map =  np.zeros((size, size), dtype = float)

    for (i,j), val in np.ndenumerate(power_map):

        k1 = i - size/2.0
        k2 = j - size/2.0
        k_map[i, j] = (np.sqrt(k1*k1 + k2*k2))

        if k_map[i,j]==0:
            power_map[i, j] = 1e-15
        else:
            power_map[i, j] = power_function(k_map[i, j])
            
    power_sample = np.random.normal(loc=0.0, scale=np.sqrt(0.5 * np.reshape(power_map, -1))) +\
               1j *  np.random.normal(loc=0.0, scale=np.sqrt(0.5 *  np.reshape(power_map, -1)))
    power_sample = np.reshape(power_sample, (size, size))
    
        # enforce hermitian and iFFT to get map
    map_fourier = np.zeros((size,size), dtype = complex)
    for (i, j,), value in np.ndenumerate(map_fourier):
        n = i - int(size/2)
        m = j - int(size/2)

        if n == 0 and m ==0:
            map_fourier[i,j] = 1e-15

        elif i ==0 or j == 0:
            map_fourier[i,j] = np.sqrt(2)*power_sample[i,j] 

        else:
            map_fourier[i,j] = power_sample[i,j] #
            map_fourier[size - i,size - j] = np.conj(power_sample[i,j])

    return  np.fft.ifft2(np.fft.fftshift(map_fourier)), power_map
```

```python
check_size = 16
number = 10000
cov = np.zeros((check_size*check_size, check_size*check_size), dtype = complex)
for count in range(number):
    map_here, power_map_here = make_map(check_size)
    map_fourier_recon_here = np.fft.fftshift(np.fft.fft2(map_here.real))
    fourier_vector = np.reshape(map_fourier_recon_here, -1)
    cov += 1.0/float(number) * np.outer(fourier_vector, np.conj(fourier_vector))
```

```python
plt.subplot(1,2,1), plt.imshow(cov.real, interpolation = 'none', cmap = cmap), plt.colorbar()
plt.subplot(1,2,2), plt.imshow(cov.imag, interpolation = 'none', cmap = cmap), plt.colorbar()
```

# Add noise to map

```python
def noise_function(i,j, sigma = 1.0):
    r_corner = np.sqrt(i*i + j*j)
    return (r_corner  + 1.0)* sigma + 0.1
```

```python
def mask_function(i,j, size):
    
    n = i - int(size/2)
    m = j - int(size/2)
    # in edges?
    if np.abs(n) > size/2.0 - size / 12.0:
        return float('nan')
    elif np.abs(m) > size/2.0 - size / 12.0:
        return float('nan')
    elif ((i > 4.6*size/7.0) and (i < 5.0*size/7.0)):
        return float('nan')
    
    circle_centre = size / 4.0
    circle_radius = size / 12
    r = np.sqrt((i - circle_centre)**2.0 + (j - circle_centre)**2.0)
    if r < circle_radius:
        return float('nan')
    
    return 0.0
```

```python
noise_variance_map = np.zeros((size,size), dtype = float)
noise_map = np.zeros((size,size), dtype = float)
mask_map = np.zeros((size,size), dtype = float)

for (i,j,), val in np.ndenumerate(noise_variance_map):
    noise_variance_map[i,j] = noise_function(i,j, 5e-3)
    mask_map[i,j] = mask_function(i,j, size)
    
noise_map = np.random.normal(loc=0.0, scale=np.sqrt(np.reshape(noise_variance_map, -1)))
noise_map = np.reshape(noise_map, (size,size))

data_no_mask = noise_map + map_pixel.real

data_mask = np.copy(data_no_mask)
noise_variance_masked = np.copy(noise_variance_map)

for (i,j,), val in np.ndenumerate(mask_map):
    if np.isnan(val) == True:
        data_mask[i,j] = 1e-15
        noise_variance_masked[i,j] = 1e15
```

# Apply messenger field

# Now do messenger field:
## Iteratively
1. $ \mathbf{t} = (\lambda \mathbf{T}) (\bar{\mathbf{N}} + \lambda \mathbf{T})^{-1} \mathbf{d} + \bar{\mathbf{N}} (\bar{\mathbf{N}} +  \lambda \mathbf{T})^{-1} \mathbf{s} $

2. $\tilde{\mathbf{s}} = \tilde{\mathbf{S}} (\tilde{\mathbf{S}} + \lambda \tilde{\mathbf{T}})^{-1} \tilde{\mathbf{t}}$

where $\bar{\mathbf{N}} = \mathbf{N} - \mathbf{T}$,  $\mathbf{T} = \tau \mathbf{1}$ and $\tau = min(diag(\mathbf{N}))$


```python
tau = np.min(noise_variance_masked)
T_diag = tau * np.ones(size*size)
Tft_diag = np.ones(size*size) * tau*float(size*size)
```

```python
N_diagonal = np.reshape(noise_variance_masked, -1)
Sft_diagonal = np.reshape(power_map, -1)
```

```python
t_data_operation = T_diag / N_diagonal
t_signal_operation = (N_diagonal - T_diag) / (N_diagonal)
Sft_plus_Tft_inv = Sft_diagonal / (Sft_diagonal + Tft_diag)
```

```python
number_of_iterations = 5000
s = np.zeros(size*size)

data_vector = np.reshape(data_mask, -1)

variance_list = []

for iteration_number in range(number_of_iterations):
    
    t = t_data_operation * data_vector +  t_signal_operation * s
    t = np.reshape(np.fft.fftshift(np.fft.fft2(np.reshape(t, (size,size)))), -1)
    s = Sft_plus_Tft_inv * t
    s = np.reshape(np.fft.ifft2(np.fft.fftshift(np.reshape(s, (size,size)))), -1)
    variance_list.append(np.var(s))

map_wiener = np.reshape(s.real, (size,size))
```

# Sample from the Wiener posterior

## $\sigma_t^2 = T \bar{N} / (T + \bar{N}) = T (N - T) / N$
## $\sigma_s^2 = \hat{S}\hat{T} / (\hat{S} + \hat{T}) $

```python
sigma_t = np.sqrt((T_diag * (N_diagonal - T_diag)) / N_diagonal)
sigma_s = np.sqrt(T_diag * Sft_diagonal / (Sft_diagonal + Tft_diag) )
```

```python
number_of_iterations = 104000
burn_in = 4000
s = np.zeros(size*size)

data_vector = np.reshape(data_mask, -1)

mean_map = np.zeros(size*size, dtype = complex)
var_map = np.zeros(size*size,  dtype = complex)

for iteration_number in range(number_of_iterations):
    
    t = t_data_operation * data_vector +  t_signal_operation * s + np.random.normal(0.0,1.0, (size*size))*sigma_t
    t = np.reshape(np.fft.fftshift(np.fft.fft2(np.reshape(t, (size,size)))), -1)
    s = Sft_plus_Tft_inv * t + np.random.normal(0.0,1.0, (size*size))*sigma_s
    s = np.reshape(np.fft.ifft2(np.fft.fftshift(np.reshape(s, (size,size)))), -1) 
    
    
    if iteration_number > burn_in:
        var_map += (1.0 / float(number_of_iterations - burn_in)) *s*s
        mean_map += (1.0 / float(number_of_iterations - burn_in)) * s

var_map -= mean_map**2.0
var_map = np.reshape(var_map, (size,size))
mean_map = np.reshape(mean_map, (size,size))
```
